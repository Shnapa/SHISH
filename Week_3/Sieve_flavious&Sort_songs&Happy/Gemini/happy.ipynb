Розробити модуль happy для пошуку щасливих трамвайних квитків. Трамвайний квиток вважається щасливим, якщо сума першої половини цифр його номера дорівнює сумі цифр його другої половини за умови що цифри сумуються до тих пір поки сумою не буде цифра. Номери квитків це восьмирозрядні числа. Якщо номер квитка містить меншу кількість цифр, ніж кількість розрядів, то номер доповнюється нулями згідно з вказаної кількості розрядів. Наприклад,  при кількості розрядів 8: 123458↦00012345

Модуль повинен містити наступні функції:
happy_number(num) для визначення чи квиток щасливий; num - натуральне число. Функція повинна повертати булеве значення True або False.

count_happy_numbers(n) для визначення кількості щасливих квитків, в діапазоні чисел від 1 до   n
 включно. Функція повинна повертати число типу int.
happy_numbers(m, n) для побудови списку щасливих квитків в заданому проміжку номерів квитків; m - нижня границя номерів квитків (включно), n - верхня границя номерів квитків (включно). Функція повинна повертати список чисел типу int.

Приклади виклику функцій:
            
happy_number(43211234)

True

happy_number(1)

False

count_happy_numbers(10001)

1

count_happy_numbers(10100)

12

happy_numbers(10001, 10010)

[10001, 10010]

happy_numbers(1, 3)

[]      

Наша імплементація

```python
'''
Happy numbers functions
'''
def happy_number(number: int) -> bool:
    '''
    int -> bool

    Returns if the number is happy

    >>> happy_number(43211234)
    True
    >>> happy_number(1)
    False
    >>> happy_number(159123)
    True
    '''
    new_number = str(number)
    counter = 0
    res1 = 0
    res2 = 0
    while len(new_number) != 8:
        new_number = '0' + new_number

    for i in str(new_number):
        counter += 1
        if counter <= 4:
            res1 += int(i)
        elif counter <= 8:
            res2 += int(i)

    while True:
        if len(str(res1)) != 1 or len(str(res2)) != 1:
            sum1, sum2 = 0, 0
            for item in str(res1):
                sum1 +=  int(item)
            for item_ in str(res2):
                sum2 += int(item_)
            res1 = sum1
            res2 = sum2
        else:
            return res1 == res2


def count_happy_numbers(n: int) -> int:
    '''
    int -> int
    
    Counts happy numbers
    
    >>> count_happy_numbers(10001)
    1
    >>> count_happy_numbers(10100)
    12
    >>> count_happy_numbers(100000)
    9999
    '''
    counter = 0
    for i in range(1, n + 1):
        if happy_number(i) is True:
            counter += 1
    return counter

def happy_numbers(m: int, n: int) -> list:
    '''
    (int, int) -> list

    Returns list of happy numbers from m to n

    >>> happy_numbers(10001, 10010)
    [10001, 10010]
    >>> happy_numbers(1, 3)
    []
    '''
    lst = []
    for i in range(m, n + 1):
        if happy_number(i) is True:
            lst.append(i)
    return lst
```
