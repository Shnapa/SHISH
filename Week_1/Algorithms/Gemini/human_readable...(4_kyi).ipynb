Your task in order to complete this Kata is to write a function which formats a duration, given as a number of seconds, in a human-friendly way.

The function must accept a non-negative integer. If it is zero, it just returns "now". Otherwise, the duration is expressed as a combination of years, days, hours, minutes and seconds.

```python
def format_duration(seconds):
    """
    Formats a duration in seconds into a human-friendly string.

    Args:
        seconds: The number of seconds to format.

    Returns:
        A human-friendly string representing the duration.
    """
    if seconds == 0:
        return "now"

    periods = [
        ("year", 31536000),
        ("day", 86400),
        ("hour", 3600),
        ("minute", 60),
        ("second", 1),
    ]

    strings = []
    for period, seconds_per_period in periods:
        if seconds >= seconds_per_period:
            count, seconds = divmod(seconds, seconds_per_period)
            if count > 0:
                strings.append(f"{count} {period}{'s' if count > 1 else ''}")

    if len(strings) > 1:
        return ", ".join(strings[:-1]) + " and " + strings[-1]
    else:
        return strings[0] if strings else "now"
```
