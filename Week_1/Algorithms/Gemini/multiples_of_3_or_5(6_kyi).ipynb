If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Finish the solution so that it returns the sum of all the multiples of 3 or 5 below the number passed in.

Additionally, if the number is negative, return 0.

Note: If the number is a multiple of both 3 and 5, only count it once.

```python
def sum_of_multiples(number):
  """
  Returns the sum of all the multiples of 3 or 5 below the number passed in.

  Args:
      number: The upper limit (exclusive) for finding multiples.

  Returns:
      The sum of all multiples of 3 or 5 below the number.
  """
  if number <= 0:
    return 0

  # Get all multiples of 3 less than the number
  threes = range(3, number, 3)
  # Get all multiples of 5 less than the number
  fives = range(5, number, 5)

  # Combine and remove duplicates using set union
  all_multiples = set(threes).union(set(fives))

  # Return the sum of all multiples
  return sum(all_multiples)
```
