Your task in order to complete this Kata is to write a function which formats a duration, given as a number of seconds, in a human-friendly way.

The function must accept a non-negative integer. If it is zero, it just returns "now". Otherwise, the duration is expressed as a combination of years, days, hours, minutes and seconds.

```python

def format_duration(seconds):
    if seconds == 0:
        return "now"

    # Define the time units and their respective seconds
    time_units = [('year', 365 * 24 * 60 * 60),
                  ('day', 24 * 60 * 60),
                  ('hour', 60 * 60),
                  ('minute', 60),
                  ('second', 1)]

    # Initialize an empty list to store formatted time components
    components = []

    # Iterate through time units
    for unit, unit_seconds in time_units:
        if seconds >= unit_seconds:
            # Calculate the number of units
            num_units = seconds // unit_seconds
            # Append the formatted string to the components list
            components.append('{} {}{}'.format(num_units, unit, 's' if num_units != 1 else ''))
            # Update the remaining seconds
            seconds %= unit_seconds

    # Join the components with commas and 'and' appropriately
    if len(components) > 1:
        return ', '.join(components[:-1]) + ' and ' + components[-1]
    else:
        return components[0]
```
