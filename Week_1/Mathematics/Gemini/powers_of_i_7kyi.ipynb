i is the imaginary unit, it is defined by i²=−1, therefore it is a solution to x²+1=0.

Your Task
Complete the function pofi that returns i to the power of a given non-negative integer in its simplest form, as a string (answer may contain i).

Тоді після пояснення і тестів, які не проходять, Gemini допрацював своє рішення

Кінцевий розв'язок

```python
def pofi(n):
    """
    This function returns i to the power of any integer, including negative powers.

    Args:
        n: An integer (positive, negative, or zero).

    Returns:
        A string representing i raised to the power of n.
    """

    if n == 0:
        return "1"  # i^0 = 1
    remainder = n % 4
    if remainder == 0:
        return "1"
    elif remainder == 1:
        return "i"
    elif remainder == 2:
        return "-1"
    else:
        return "-i"
```
