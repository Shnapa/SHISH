DESCRIPTION:
i is the imaginary unit, it is defined by i²=−1, therefore it is a solution to x²+1=0.

Your Task
Complete the function pofi that returns i to the power of a given non-negative integer in its simplest form, as a string (answer may contain i).

Оскільки умова завдання сама по собі є короткою, використовує зрозуміле математиці поняття, то навіть з таким вичерпним описом chat GPT дав правильний розв'язок одразу

Розв'язок

```python
def pofi(n):
    # Reduce the power modulo 4 to get one of the four possible values
    power = n % 4
    
    # Define the values of i^0, i^1, i^2, i^3
    values = [1, 'i', -1, '-i']
    
    # Return the appropriate value based on the reduced power
    return str(values[power])

# Test cases
print(pofi(0))  # Output: '1'
print(pofi(1))  # Output: 'i'
print(pofi(2))  # Output: '-1'
print(pofi(3))  # Output: '-i'
print(pofi(4))  # Output: '1'
print(pofi(5))  # Output: 'i'
# and so on...
```
