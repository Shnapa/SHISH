```python
import math

class Sphere:
  """
  Represents a sphere with radius and mass properties.
  """
  def __init__(self, radius, mass):
    """
    Initializes a Sphere object with radius and mass.

    Args:
      radius: The radius of the sphere (float).
      mass: The mass of the sphere (float).
    """
    self.radius = radius
    self.mass = mass

  def get_radius(self):
    """
    Returns the radius of the sphere.

    Returns:
      The radius of the sphere (float).
    """
    return self.radius

  def get_mass(self):
    """
    Returns the mass of the sphere.

    Returns:
      The mass of the sphere (float).
    """
    return self.mass

  def get_volume(self):
    """
    Calculates and returns the volume of the sphere rounded to 5 decimal places.

    Returns:
      The volume of the sphere (float) rounded to 5 decimal places.
    """
    volume = (4/3) * math.pi * (self.radius**3)
    return round(volume, 5)

  def get_surface_area(self):
    """
    Calculates and returns the surface area of the sphere rounded to 5 decimal places.

    Returns:
      The surface area of the sphere (float) rounded to 5 decimal places.
    """
    surface_area = 4 * math.pi * (self.radius**2)
    return round(surface_area, 5)

  def get_density(self):
    """
    Calculates and returns the density of the sphere rounded to 5 decimal places.

    Returns:
      The density of the sphere (float) rounded to 5 decimal places.
    """
    if self.radius == 0:
      # Avoid division by zero error
      return 0
    density = self.mass / ((4/3) * math.pi * (self.radius**3))
    return round(density, 5)

# Example usage
ball = Sphere(2, 50)
print(f"Radius: {ball.get_radius()}")
print(f"Mass: {ball.get_mass()}")
print(f"Volume: {ball.get_volume()}")
print(f"Surface Area: {ball.get_surface_area()}")
print(f"Density: {ball.get_density()}")
```
